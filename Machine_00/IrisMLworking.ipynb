```python
sepal_l=float(input("Enter Sepal-Length : "))
sepal_w=float(input("Enter Sepal-Wedgth : "))
petal_l=float(input("Enter Petal-Length : "))
petal_w=float(input("Enter Petal-Wedgth : "))

import ipywidgets as widgets
from IPython.display import display
button = widgets.Button(description="Predict Class")
output = widgets.Output()

display(button, output)

def on_button_clicked(b):
    with output:
        name=HighM.predict([[sepal_l,sepal_w,petal_l,petal_w]])
        print('IRIS flower class is:',name[0])

button.on_click(on_button_clicked)
```
